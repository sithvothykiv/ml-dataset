```python
import numpy as np
```

```python
X = [
    [148, 24, 1385],
    [132, 25, 2031],
    [453, 11, 86],
    [158, 24, 105],
    [172, 25, 201],
    [413, 11, 86.0002222],
    [38,  54, 185],
    [142, 25, 431],
    [453, 31, 86.001],
]
X = np.array(X)
X.shape, X
```

```python
ones = np.ones(X.shape[0])
X = np.column_stack([ones, X])
y = [10000, 20000, 15000, 20050, 10000, 20000, 15000, 25000, 12000]
```

```python
XTX = X.T.dot(X)
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
w_full = XTX.dot(X.T).dot(y)
w_full
```

```python
w0 = w_full[0]
w = w_full[1:]
w0, w
```

```python
def train_linear_regression(X, y):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)

    return w_full[0], w_full[1:]
```

```python
train_linear_regression(X, y)
```

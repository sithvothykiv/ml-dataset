## 5. Example: Predicting final score (G3) using Pandas logic

```python
df["average_score"] = (df["G1"] + df["G2"]) / 2
```

## 6. Simple performance prediction (rule-based)

```python
df["predicted_performance"] = np.where(
    df["average_score"] >= 10, "Pass", "Fail"
)
```

## 7. Analyze results

```python
df["predicted_performance"].value_counts()
```

## 8. Groupby analysis (important for marks)

```python
# Performance by gender
df.groupby("sex")["G3"].mean()

# Performance by study time
df.groupby("studytime")["G3"].mean()
```

## 9. Save the new dataset

```python
df.to_csv("Rithy_Student_performance_with_prediction.csv", index=False)
```

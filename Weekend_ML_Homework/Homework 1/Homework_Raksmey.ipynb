### Q2. Record Count

### Q3. Laptop brands

```python
lap_brd = df["Brand"].nunique()

print("Total Laptop brands is:", lap_brd)
```

### Q4. Missing values

```python
missing_columns = df.isnull().any().sum()

print("Missing value is:", missing_columns)
```

### Q5. Maximum final price

```python
dell_notebooks = df[df["Brand"] == "Dell"]
max_price = dell_notebooks["Final Price"].max()

print("Maximum final price of Dell notebooks:", max_price)
```

### Q6. Median value of Screen

```python
# 1. Find the median value of Screen column in the dataset.
median_screen = df["Screen"].median()
print("The Median value of Screen column is:", median_screen)
```

```python
# 2. Next, calculate the most frequent value of the same Screen column
most_frequent = df["Screen"].mode()[0]

print("The Most frequent value in Screen column is:", most_frequent)
```

```python
# 3. Use fillna method to fill the missing values in Screen column with the most frequent value from the previous step.
print("Columns missing value before is:", df["Screen"].isna().sum())

df["Screen"] = df["Screen"].fillna(df["Screen"].mode()[0])
missing_val_after = df["Screen"].isna().sum()
print("Columns missing value After is:", missing_val_after)
```

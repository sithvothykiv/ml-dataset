### Q6. Median value of Screen

```python
# 1. Find the median value of Screen column in the dataset.
medianscreen = df["Screen"].median()
print("The Median value of Screen column is:", medianscreen)
```

```python
# 2. Next, calculate the most frequent value of the same Screen column
mostfrequent = df["Screen"].mode()[0]
print("The Most frequent value in Screen column is:", mostfrequent)
```

```python
# 3. Use fillna method to fill the missing values in Screen column with the most frequent value from the previous step.
print("Columns missing value before is:", df["Screen"].isna().sum())

df["Screen"] = df["Screen"].fillna(df["Screen"].mode()[0])
missingvalafter = df["Screen"].isna().sum()
print("Columns missing value After is:", missingvalafter)
```

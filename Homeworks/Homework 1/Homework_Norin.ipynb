```python
# Q2. Records count
#How many records are in the dataset?

linkCSV = 'https://raw.githubusercontent.com/sithvothykiv/ml-laptop-dataset/refs/heads/main/laptops.csv'

df = pd.read_csv(linkCSV)
print(f"Records Count: {len(df)}")

df.shape
```

```python
# Q3. Laptop brands
# How many laptop brands are presented in the dataset?

brands_count = df['Brand'].nunique()
print(f"Laptop Brands Count: {brands_count}")
```

```python
# Q4. Missing values
# How many columns in the dataset have missing values?

missing_values = df.columns[df.isnull().any()].tolist()
missing_values_count = len(missing_values)

print(f"Total Columns in the Dataset with Missing Values: {missing_values_count}")
```

```python
# Q5. Maximum final price
# What's the maximum final price of Dell notebooks in the dataset?

dell_notebooks = df[df['Brand'] == 'Dell']
max_final_price_dell = dell_notebooks['Final Price'].max()

print(f"Maximum Final Price of Dell Notebooks: {max_final_price_dell} $")
```

```python
# Q6. Median value of Screen
# 1. Find the median value of Screen column in the dataset.
# 2. Next, calculate the most frequent value of the same Screen column.
# 3. Use fillna method to fill the missing values in Screen column with the most frequent value from the previous step.
# 4. Now, calculate the median value of Screen once again.

# Has it changed?
# Hint: refer to existing mode and median functions to complete the task.
# No/Yes

# Answer 1
median_screen_before = df['Screen'].median()
print(f"1/ Median value of Screen column in dataset : {median_screen_before}")

# Answer 2
most_frequent_screen = df['Screen'].mode()[0]
print(f"2/ Most frequent value of Screen column : {most_frequent_screen}")

# Answer 3
missing_value_fill = df['Screen'].fillna(most_frequent_screen)
print(f"3/ Missing values in Screen column filled with most frequent value from previous step: {missing_value_fill}")

# Answer 4
median_screen_after = df['Screen'].median()
print(f"4/ Median value of Screen re-calculate : {median_screen_after}")

# Has it changed?
print("Has it changed? No")


```
